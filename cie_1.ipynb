```python
import pandas as pd
df=pd.read_csv("/content/iris-write-from-docker.csv")
print(df)
df.info()
df.tail()
df.head()
df.describe()
```

```python
import matplotlib.pyplot as plt
species_counts = df['class'].value_counts()
plt.bar(species_counts.index, species_counts.values, color='lightblue')
plt.title("Number of Flowers by Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.grid(axis='y')
```

```python
import matplotlib.pyplot as plt
plt.hist(df['petal_length'], bins=10, color='lightgreen', edgecolor='black')
plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['petal_length'], df['petal_width'], color='purple', alpha=0.6)
plt.title("Petal Length vs Petal Width")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.show()
```

```python
import matplotlib.pyplot as plt
species = df['class'].unique()
data = [df[df['class'] == s]['petal_length'] for s in species]
plt.boxplot(data, labels=species)
plt.title("Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.grid(True)
plt.show()
```
